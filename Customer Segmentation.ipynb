```python
df.drop(['Genre'], axis = 1, inplace = True)
```

```python
from sklearn.model_selection import train_test_split 
x = df.drop(['Annual Income (k$)'],axis=1) 
y = df[['Annual Income (k$)']] 
```

```python
x_train,x_test, y_train, y_test = train_test_split(x,y,random_state=104,test_size=0.25,shuffle=True) 
```

```python
from sklearn import linear_model

logr = linear_model.LinearRegression()
logr.fit(x_train,y_train)
```

```python
y_predicted = logr.predict(x_test)
```

```python
from sklearn.metrics import r2_score 
```

```python
r2 = r2_score(y_test,y_predicted) 
```

```python

print('Coefficient of determination: ',r2*100)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor 
from sklearn.metrics import r2_score, mean_absolute_error
import numpy as np
from sklearn import linear_model
import warnings
warnings.filterwarnings('ignore')
```

```python
(y, x) = (df['CustomerID'].values, df.drop(['CustomerID'], axis = 1))
(x_train, x_test, y_train, y_test) = train_test_split(x, y, test_size = 0.3)
```

```python
logr = linear_model.LogisticRegression()
logr.fit(x_train, y_train)
lr_train_score = round(logr.score(x_train, y_train), 2)
lr_test_score = round(logr.score(x_test, y_test), 2)

y_pred_lr = logr.predict(x_test)

print('Coefficient of determination: ', r2_score(y_test, y_pred_lr)*100)
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_train, y_train)


nb_train_score = round(nb.score(x_train, y_train), 2)
nb_test_score = round(nb.score(x_test, y_test), 2)

y_pred_lr = nb.predict(x_test)


print('Coefficient of determination: ', r2_score(y_test, y_pred_lr)*100)
```
